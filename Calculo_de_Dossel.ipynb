# Cálculo da Transmitância da Radiação no Interior do Dossel
1. Fórmula da Transmitância no Dossel

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo ângulos zenitais de 0 a 85 graus com passos de 5 graus
theta = np.arange(0, 86, 5)  # em graus
theta_rad = np.radians(theta)  # convertendo para radianos

# Índices de Área Foliar (IAF)
IAF_values = [1, 5]

# Distribuição de Folhas: G(theta)
def G_horizontal(theta):
    return np.cos(theta)

def G_vertical(theta):
    return 1

def G_spherical(theta):
    return 0.5

# Cálculo da transmitância T para diferentes distribuições de folhas
def calculate_transmittance(G_function, IAF):
    return np.exp(-G_function(theta_rad) * IAF / np.cos(theta_rad))

# Plotando os gráficos de transmitância para cada IAF
for IAF in IAF_values:
    plt.figure(figsize=(10, 6))

    T_horizontal = calculate_transmittance(G_horizontal, IAF)
    T_vertical = calculate_transmittance(G_vertical, IAF)
    T_spherical = calculate_transmittance(G_spherical, IAF)

    plt.plot(theta, T_horizontal * 100, label='Distribuição Horizontal')
    plt.plot(theta, T_vertical * 100, label='Distribuição Vertical')
    plt.plot(theta, T_spherical * 100, label='Distribuição Esférica')

    plt.xlabel('Ângulo Zenital (°)')
    plt.ylabel('Transmitância (%)')
    plt.title(f'Transmitância da Radiação no Interior do Dossel para IAF = {IAF}')
    plt.legend()
    plt.grid(True)

    # Ensure the directory exists before saving
    #!mkdir -p /content/drive/MyDrive/Colab\ Notebooks/Data_meteo/ResultDossel/

    if IAF == 1:
        plt.savefig(f'/content/drive/MyDrive/Colab Notebooks/Data_meteo/Result_images/transmittance_IAF_{IAF}_distribuição.png')

    if IAF == 5:
        plt.savefig(f'/content/drive/MyDrive/Colab Notebooks/Data_meteo/Result_images/transmittance_IAF_{IAF}_distribuição.png')

    plt.show()
```
